```python
region_retailer_sales = df.groupby(['Region', 'Retailer'])['Total Sales'].sum()

```

```python
# this gives total sales done by each retailer in each region
region_retailer_sales
```

```python
region_retailer_sales.index.get_level_values('Region').unique()
```

```python
region_retailer_sales.loc['West'].index.get_level_values('Retailer')
```

```python
region_retailer_sales.values
```

```python


# Iterate over each region
for region in region_retailer_sales.index.get_level_values('Region').unique():
    region_data = region_retailer_sales.loc[region]

    # Create a bar chart for the region
    plt.figure()
    plt.bar(region_data.index.get_level_values('Retailer'), region_data.values)
    plt.xlabel('Retailer')
    plt.ylabel('Total Sales')
    plt.title(f'Total Sales by Retailer in {region}')
    plt.show()
```

```python
state_region_sales=df.groupby(['Region','State'])['Total Sales'].sum()
```

```python
# this gives state wise total sales in region.
state_region_sales
```

```python
state_region_sales.index.get_level_values('Region').unique()
```

```python
for region in state_region_sales.index.get_level_values('Region').unique():
  region_data=state_region_sales.loc[region]
  plt.figure()
  plt.barh(region_data.index.get_level_values('State'),region_data.values)
  plt.xlabel('State')
  plt.ylabel('Total Sales')
  plt.title(f'Total Sales by state in {region}')
  plt.show()
```

```python
retail=pd.crosstab(df['Region'],df['Retailer'])
retail
# this tab provide the information about the total sales done by the retailer which is greater than the other.
# this is because of number of retailer present in that region.
```

```python
for idx,row in retail.iterrows():
  plt.figure()
  row.plot.pie(autopct='%0.1f%%')
  plt.axis('equal')
  plt.show()
```

```python
cross_tab=pd.crosstab(df['Retailer'],df['Sales Method'])
# from this ,we conclude that retailer sale their product through online  as compare to other sales method except Walmart.
cross_tab
```
